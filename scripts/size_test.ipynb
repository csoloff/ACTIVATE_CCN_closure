```python
from closure_funcs import *
import numpy as np
from scipy.optimize import root_scalar
from matplotlib import pyplot as plt
import pandas as pd
from tqdm import tqdm
from joblib import Parallel, delayed
from scipy.stats import linregress
import matplotlib.colors as mcolors
import matplotlib as mpl
```

```python
x = np.logspace(-10, -1, 100)
```

```python
y = []
for i in range(0,len(x)):
    y.append(find_d_act(.37, x[i]))
```

```python
mpl.rcParams['mathtext.fontset'] = 'stix'
fig, ax = plt.subplots(figsize=(4,3))
ax.plot(x, y, c='black', ls='-', label = 's = 0.37%')
ax.axhline(y[0], c='red', ls='--', alpha=.8, label='$\mathsf{D}_\mathsf{p,c}$ = ' + str(round(y[0],1))+' nm')
ax.set_xscale('log')
ax.set_ylabel('$\mathsf{D}_\mathsf{p,c}$ [nm]', fontsize=12)
ax.set_xlabel('$\kappa$', fontsize=15)
ax.legend(fontsize=12)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
plt.savefig('../figures/d_act_vs_kappa.png', dpi=300, bbox_inches='tight')
```
